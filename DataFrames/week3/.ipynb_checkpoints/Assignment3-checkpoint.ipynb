```python
def answer_one():
    import pandas as pd
    global finalresult,Energy,ScimEn
    Energy=pd.read_excel('assets/Energy Indicators.xls')
    Energy=Energy[17:244]
    Energy.drop('Unnamed: 0',inplace=True,axis='columns')
    Energy.drop('Unnamed: 1',inplace=True,axis='columns')
    Energy.rename(columns={'Unnamed: 2':'Country','Unnamed: 3':'Energy Supply','Unnamed: 4':'Energy Supply per Capita',
                  'Unnamed: 5':'% Renewable'},inplace=True)
    Energy['Energy Supply']=Energy['Energy Supply'].apply(lambda x:convert_gegajoul(x))
    Energy=Energy.reset_index()
    Energy.drop('index',inplace=True,axis='columns')
    Energy=Energy.set_index('Country')
    Energy=Energy.rename(index={'Republic of Korea':'South Korea','United States of America':'United States',
                     'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
                     'China, Hong Kong Special Administrative Region':'Hong Kong'})
    Energy=Energy.reset_index()
    Energy['Country']=Energy['Country'].apply(lambda x: rename_country(x))
    Energy=Energy.set_index('Country')
    
    #Reading the second file
    GDP=pd.read_csv('assets/world_bank.csv')
    GDP=GDP.rename(columns=GDP.iloc[3])
    GDP=GDP[4:-1]
    GDP=GDP.reset_index()
    GDP.drop('index',inplace=True,axis='columns')
    for element in GDP.columns[4:len(GDP.columns)]:
        GDP=GDP.rename(columns={element:int(element)})
    GDP.rename(columns={'Country Name':'Country'},inplace=True)
    GDP=GDP.set_index('Country')
    GDP.rename(index={'Korea, Rep.':'South Korea','Iran, Islamic Rep.':'Iran',
                     'Hong Kong SAR, China':'Hong Kong'
                     },inplace=True)
    #Read the third file
    ScimEn=pd.read_excel('assets/scimagojr-3.xlsx')
    ScimEn=ScimEn.set_index('Country')
    newgdp=GDP[[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]].copy()
    result=pd.merge(ScimEn,newgdp,how='inner',left_index=True,right_index=True)
    finalresult=pd.merge(result,Energy,how='left',left_index=True,right_index=True)
    return finalresult[0:15]
```
